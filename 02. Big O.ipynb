```
O(n ^ 2) O of n square
One for loop inside of the other for loop
def print_item(n):
    for i in range(n):
        for j in range(n):
            print(i, j)

We'll get a very diffrent result when we have loop inside of a loop.

So when we printed to this to the screen, we got n times n items printing out.

n * n = n^2
O(n^2)

So instead of doing this as two, for loops, one after the other, let's look at what happens when
we take this one level deeper.
def print_item(n):
    for i in range(n):
        for j in range(n):
            for k in range(n):
                print(i, j, k)

Now we have for loop within a for loop with in a for loop.
this ran n * n * n = n^3
O(n^3)
So doesn't matter if it's O(n) to the 3rd, 4th, 10th, we're going to simplify it as O of n^2.

So O of n ^ 2 is much steeper than O of n.

Lot less efficient from a time complexity standpoint.

```

````python
def print_item(n):
    for i in range(n):
        for j in range(n):
            print(i, j)
print(print_item(10))
````

````python
# Take these one level deeper 
def print_item(n):
    for i in range(n):
        for j in range(n):
            for k in range(n):
                print(i, j, k)
print(print_item(10))
````

```
Drop Non-Dominants - simplification technique for Big O is called drop non-dominants.

def print_items(n):
    for i in range(n):
        for j in range(n):  # Nested for loop ran O(n^2) times
            print(i, j)
                                          And the total number of items that we output was O of n^2 pluse O of n so we could write it like O(n^2 + n)
    for k in range(n):
        print(k)      # for loop ran O(n)
print_items(10)

What happended when we start having n be a very large
And as the number becomes 1000 or a million that stand alone n is vary
small portion of the number, so in this equation n^2 is the dominant term and 
that stand alone n is the non-dominant term. so n is non domainant term so we 
drop the n and O(n^2) 

```

````python
def print_items(n):
    for i in range(n):
        for j in range(n):
            print(i, j)

    for k in range(n):
        print(k)
print_items(10)
````
